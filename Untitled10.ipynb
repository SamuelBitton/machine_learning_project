## Data Preparation:
#### Basic Cleaning + Feature Engineering
We create time-based features and handle obvious invalid values / missing values.


```python
# Work on a clean copy
df_new = df.copy()

# Convert date strings to datetime
df_new["DATE OCC"]  = pd.to_datetime(df_new["DATE OCC"],  format="%m/%d/%Y %I:%M:%S %p", errors="coerce")
df_new["Date Rptd"] = pd.to_datetime(df_new["Date Rptd"], format="%m/%d/%Y %I:%M:%S %p", errors="coerce")


# Create temporal features from occurrence date
df_new["occ_month"] = df_new["DATE OCC"].dt.month
df_new["occ_day"]   = df_new["DATE OCC"].dt.day
df_new["occ_dow"]   = df_new["DATE OCC"].dt.dayofweek


# Convert TIME OCC (HHMM) into hour/minute features with validation
t = pd.to_numeric(df_new["TIME OCC"], errors="coerce")
hh = t // 100
mm = t % 100

valid_time = t.between(0, 2359) & hh.between(0, 23) & mm.between(0, 59)
df_new.loc[~valid_time, "TIME OCC"] = np.nan

t2 = pd.to_numeric(df_new["TIME OCC"], errors="coerce")
df_new["occ_hour"] = (t2 // 100)
df_new["occ_min"]  = (t2 % 100)


# Replace invalid victim ages with NaN (do not drop rows)
age = pd.to_numeric(df_new["Vict Age"], errors="coerce")
df_new["Vict Age"] = age.mask((age <= 0) | (age > 100), np.nan)

# Fill missing categorical values with a placeholder category
for c in ["Vict Sex", "Vict Descent"]:
    if c in df_new.columns:
        df_new[c] = df_new[c].astype("string").str.strip().fillna("Unknown")


# Drop columns with excessive missing values
cols_to_drop = ["Crm Cd 2", "Crm Cd 3", "Crm Cd 4", "Weapon Used Cd", "Weapon Desc", "Cross Street", "Mocodes"]
cols_to_drop = [c for c in cols_to_drop if c in df_new.columns]

df_new = df_new.drop(columns=cols_to_drop)

print("After cleaning:", df_new.shape)
```

We explicitly removed all crime code columns (Crm Cd and Crm Cd 1–4) to prevent label leakage, since these variables map almost deterministically to the target variable Crm Cd Desc. Keeping them would allow the model to trivially infer the label instead of learning meaningful spatio-temporal and contextual patterns.

#### Preprocessing Pipeline (numeric + categorical)

We build a preprocessing pipeline that:
- imputes missing values
- scales numeric features
- one-hot encodes categorical features

This ensures consistent preprocessing across all classification models.


```python
def make_sklearn_safe(df):
    df = df.copy()
    for c in df.columns:
        if pd.api.types.is_numeric_dtype(df[c]):
            df[c] = pd.to_numeric(df[c], errors="coerce")
        else:
            df[c] = df[c].astype("object")
            df[c] = df[c].where(df[c].notna(), np.nan)
    return df

# Make X sklearn-compatible (remove pandas <NA>)
X_train_safe = make_sklearn_safe(X_train)
X_test_safe  = make_sklearn_safe(X_test)

# Split columns by type
num_cols = X_train_safe.select_dtypes(include=["number"]).columns.tolist()
cat_cols = X_train_safe.select_dtypes(exclude=["number"]).columns.tolist()

# Pipelines
numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler(with_mean=False))
])

categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", min_frequency=50))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe, num_cols),
        ("cat", categorical_pipe, cat_cols),
    ]
)

print("Numeric cols:", num_cols)
print("Categorical cols:", cat_cols)
print("Preprocessing pipeline is ready.")
```

## Classification Models:

#### Baseline Model (Quick Sanity Check)
Before training the main models, we train a fast baseline classifier to verify that:
- the preprocessing pipeline works end-to-end
- the train/test split is correct
- we have a reference performance to compare against


```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, f1_score

baseline_clf = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", SGDClassifier(loss="log_loss", max_iter=200, tol=1e-3, random_state=42))
])

baseline_clf.fit(X_train_safe, y_train)
y_pred = baseline_clf.predict(X_test_safe)

print("Baseline (SGD) - Test Results")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Macro F1:  {f1_score(y_test, y_pred, average='macro'):.4f}")
```

##### Baseline Results (SGD)
The baseline SGD classifier achieved **Accuracy = 0.4539** and **Macro F1 = 0.2318** on the test set.

The lower Macro F1 is expected in this multi-class, imbalanced setting because rare classes are harder to predict and Macro F1 penalizes poor performance on them.


### Model 1 — Random Forest (Tree-based Baseline)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import StratifiedShuffleSplit

#  Preprocessing for tree-based models (Ordinal Encoding instead of One-Hot) 

num_cols = X_train_safe.select_dtypes(include=["number"]).columns.tolist()
cat_cols = X_train_safe.select_dtypes(exclude=["number"]).columns.tolist()

numeric_pipe_tree = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_pipe_tree = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
])

preprocess_tree = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe_tree, num_cols),
        ("cat", categorical_pipe_tree, cat_cols),
    ],
    remainder="drop"
)

#  Stratified subsample for faster RF training ---

RF_TRAIN_SIZE = 200_000  
sss = StratifiedShuffleSplit(n_splits=1, train_size=RF_TRAIN_SIZE, random_state=42)
rf_train_idx, _ = next(sss.split(X_train_safe, y_train))

X_train_rf = X_train_safe.iloc[rf_train_idx]
y_train_rf = y_train.iloc[rf_train_idx]

print("RF training subset shapes:", X_train_rf.shape, y_train_rf.shape)

# Random Forest model (speed-optimized parameters) 

rf_clf = Pipeline(steps=[
    ("preprocess", preprocess_tree),
    ("model", RandomForestClassifier(
        n_estimators=200,        # fewer trees = faster
        max_depth=30,            # limit depth = much faster + less overfit
        min_samples_leaf=20,     # bigger leaves = faster + smoother
        max_features="sqrt",     # standard for RF
        class_weight="balanced", # handle imbalance
        n_jobs=-1,
        random_state=42
    ))
])

# --- 4) Train on subset, evaluate on full test set ---

%time rf_clf.fit(X_train_rf, y_train_rf)

rf_pred = rf_clf.predict(X_test_safe)

print("\nRandom Forest - Test Results (trained on stratified 200k subset)")
print(f"Accuracy: {accuracy_score(y_test, rf_pred):.4f}")
print(f"Macro F1:  {f1_score(y_test, rf_pred, average='macro'):.4f}")

# Optional: detailed per-class report (can be long)
# print(classification_report(y_test, rf_pred))
```

The Random Forest model significantly improves over the baseline, increasing Macro F1.
This indicates that non-linear interactions between spatial, temporal, and categorical features are important for predicting crime types, and that the model handles class imbalance better than the linear baseline.

```python
import numpy as np
import xgboost as xgb
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import LabelEncoder

# --- 1) Encode labels to integers ---
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc  = le.transform(y_test)
num_classes = len(le.classes_)
print("Num classes:", num_classes)

# --- 2) Stratified subset for training speed ---
XGB_TRAIN_SIZE = 300_000
sss = StratifiedShuffleSplit(n_splits=1, train_size=XGB_TRAIN_SIZE, random_state=42)
xgb_train_idx, _ = next(sss.split(X_train_safe, y_train))

X_train_xgb = X_train_safe.iloc[xgb_train_idx]
y_train_xgb = y_train_enc[xgb_train_idx]

print("XGB training subset shape:", X_train_xgb.shape)

# --- 3) One-hot preprocess (same as your linear/MLP pipeline) ---
X_train_xgb_t = preprocess.fit_transform(X_train_xgb)
X_test_xgb_t  = preprocess.transform(X_test_safe)

# --- 4) Train/valid split for early stopping ---
sss2 = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
tr_idx, va_idx = next(sss2.split(X_train_xgb_t, y_train_xgb))

X_tr, y_tr = X_train_xgb_t[tr_idx], y_train_xgb[tr_idx]
X_va, y_va = X_train_xgb_t[va_idx], y_train_xgb[va_idx]

# --- 5) Build DMatrix ---
dtrain = xgb.DMatrix(X_tr, label=y_tr)
dvalid = xgb.DMatrix(X_va, label=y_va)
dtest  = xgb.DMatrix(X_test_xgb_t)

# --- 6) Params (fast + strong) ---
params = {
    "objective": "multi:softprob",
    "num_class": num_classes,
    "learning_rate": 0.05,
    "max_depth": 8,
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    "min_child_weight": 1,
    "lambda": 1.0,
    "eval_metric": "mlogloss",
    "tree_method": "hist",   # fast on CPU
    "seed": 42,
    "verbosity": 1
}

num_boost_round = 3000
early_stopping_rounds = 50

# --- 7) Train with early stopping (THIS works even on older versions) ---
%time booster = xgb.train(params=params,dtrain=dtrain,num_boost_round=num_boost_round,evals=[(dvalid, "valid")],early_stopping_rounds=early_stopping_rounds,verbose_eval=100)

print("Best iteration:", booster.best_iteration)

# --- 8) Predict + evaluate ---
proba = booster.predict(dtest)               # shape: (n_samples, num_classes)
y_pred_enc = np.argmax(proba, axis=1)
y_pred = le.inverse_transform(y_pred_enc)

print("\nXGBoost - Test Results")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Macro F1:  {f1_score(y_test, y_pred, average='macro'):.4f}")

```

The XGBoost model achieved an accuracy of 0.5132 and a macro F1-score of 0.3158.

### MODEL 3 AdaBoost

# AdaBoost uses many weak learners (small decision trees) and combines them into a stronger model.

The original AdaBoost configuration was computationally expensive due to the large size
of the dataset (nearly one million samples) and the high-dimensional feature space created
by One-Hot Encoding.

AdaBoost trains weak learners sequentially, which makes it significantly slower than
modern gradient boosting methods on large datasets. Therefore, we intentionally simplified
the model by reducing the number of estimators and limiting the complexity of the base
learner.

This simplified configuration allows AdaBoost to serve as a reasonable benchmark model
while keeping training time manageable.

Why we test it:
- It is a classic boosting algorithm (baseline comparison vs. XGBoost).
- It can capture non-linear patterns, but often struggles more than XGBoost on very large / noisy / highly multi-class data.

For speed and fairness, we train on a stratified subset of the training set (same approach as Random Forest/XGBoost).

```python
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder

# ----------------------------
# 1) Stratified subset for speed
# ----------------------------
ADABOOST_TRAIN_SIZE = 200_000  

sss = StratifiedShuffleSplit(n_splits=1, train_size=ADABOOST_TRAIN_SIZE, random_state=42)
train_idx, _ = next(sss.split(X_train_safe, y_train))

X_train_ab = X_train_safe.iloc[train_idx]
y_train_ab = y_train.iloc[train_idx] if hasattr(y_train, "iloc") else y_train[train_idx]

print("AdaBoost training subset shape:", X_train_ab.shape)

# ----------------------------
# 2) Preprocessing (tree-friendly): impute + ordinal encode categoricals
# ----------------------------
num_cols = X_train_safe.select_dtypes(include=["number"]).columns.tolist()
cat_cols = X_train_safe.select_dtypes(exclude=["number"]).columns.tolist()

numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
])

preprocess_tree = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe, num_cols),
        ("cat", categorical_pipe, cat_cols),
    ],
    remainder="drop"
)

# ----------------------------
# 3) AdaBoost model (SAMME for multi-class)
# ----------------------------
base_tree = DecisionTreeClassifier(
    max_depth=2,               # depth=1 is "stump"; depth=2 often better
    min_samples_leaf=20,       # helps speed + stability
    class_weight="balanced",   # handle imbalance a bit
    random_state=42
)

ada = AdaBoostClassifier(
    estimator=base_tree,       # if you get an error, replace "estimator" with "base_estimator"
    n_estimators=400,          # 200-600 reasonable
    learning_rate=0.5,
    algorithm="SAMME",         # multi-class safe
    random_state=42
)

ab_clf = Pipeline(steps=[
    ("preprocess", preprocess_tree),
    ("model", ada)
])

# ----------------------------
# 4) Train + Evaluate
# ----------------------------
%time ab_clf.fit(X_train_ab, y_train_ab)

ab_pred = ab_clf.predict(X_test_safe)

print("\nAdaBoost - Test Results")
print(f"Accuracy: {accuracy_score(y_test, ab_pred):.4f}")
print(f"Macro F1:  {f1_score(y_test, ab_pred, average='macro'):.4f}")

# Optional (can be long)
# print(classification_report(y_test, ab_pred))
```

The AdaBoost model achieved lower performance compared to XGBoost, which is expected given
its simpler structure and the large scale of the dataset. AdaBoost was therefore used as a
benchmark boosting method rather than a state-of-the-art solution for this problem.

### Model 5 Stacking Ensemble

Stacking combines multiple different models (base learners) and trains a final meta-model that learns how to best combine their predictions.
This often improves performance because different models capture different patterns in the data.

In this project, we use:
- Base learners: a linear model (fast) + Random Forest (non-linear)
- Meta learner: Logistic Regression (simple + robust)

To control runtime, we train the stack on a stratified subset of the training set and evaluate on the full test set.

```python
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit

STACK_TRAIN_SIZE = 200_000   # you can try 300_000 if still fast

sss = StratifiedShuffleSplit(n_splits=1, train_size=STACK_TRAIN_SIZE, random_state=42)
stack_idx, _ = next(sss.split(X_train_safe, y_train))

X_train_st = X_train_safe.iloc[stack_idx]
y_train_st = y_train.iloc[stack_idx] if hasattr(y_train, "iloc") else y_train[stack_idx]

print("Stacking training subset shape:", X_train_st.shape)

# One-hot preprocessing (same preprocess as linear/MLP)
X_train_st_t = preprocess.fit_transform(X_train_st)
X_test_st_t  = preprocess.transform(X_test_safe)

print("Transformed shapes:", X_train_st_t.shape, X_test_st_t.shape)
```

### Stacking – Training

We define two base learners:
1) Linear model (Logistic Regression) – fast and strong on sparse one-hot features
2) Random Forest – captures non-linear interactions

The meta-model is Logistic Regression.
We use cross-validation inside stacking (cv=3) to generate out-of-fold predictions for the meta-learner.

```python
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

# Base learners
base_estimators = [
    ("lr", LogisticRegression(max_iter=200, n_jobs=-1, class_weight="balanced")),
    ("rf", RandomForestClassifier(
        n_estimators=200,
        random_state=42,
        n_jobs=-1,
        class_weight="balanced",
        min_samples_leaf=10
    ))
]

# Meta learner
final_estimator = LogisticRegression(max_iter=300, n_jobs=-1)

stack_clf = StackingClassifier(
    estimators=base_estimators,
    final_estimator=final_estimator,
    cv=3,
    n_jobs=-1,
    passthrough=False
)

%time stack_clf.fit(X_train_st_t, y_train_st)

stack_pred = stack_clf.predict(X_test_st_t)

print("\nStacking Ensemble - Test Results")
print(f"Accuracy: {accuracy_score(y_test, stack_pred):.4f}")
print(f"Macro F1:  {f1_score(y_test, stack_pred, average='macro'):.4f}")
```

Why didn’t stacking perform better?

Because XGBoost was already very strong and both base models are tree-based. Stacking helped balance class-level performance but did not significantly improve overall accuracy, which is a common and realistic outcome in large multi-class problems.

# Unsupervised Learning: Geo-temporal Crime Clustering (K-Means vs DBSCAN)

In the supervised section, we predicted crime categories using labeled data.
In this unsupervised section, we aim to discover natural crime patterns **without using labels during training**.

Based on our project proposal, we focus on **geospatial and temporal clustering** to identify actionable crime hot-spots and profiles:

- **K-Means**: applied to cleaned **LAT/LON** and **HOUR** features to segment the city into **N distinct crime profiles**.
- **DBSCAN**: a density-based method used to detect **irregularly shaped hot-spots** and automatically flag **low-density incidents as anomalies/outliers**.

We then interpret clusters and compare K-Means vs DBSCAN in terms of:
- cluster structure,
- hotspot detection ability,
- and anomaly detection (DBSCAN noise points).

##  Build the clustering dataset (LAT/LON + HOUR)

We construct an unsupervised feature matrix using only:
- Geographic coordinates: **LAT, LON**
- Temporal component: **occ_hour** (hour of day)

These are the core features required for geo-temporal clustering.

```python
import pandas as pd
import numpy as np

# We will use X_train_safe (no labels) for clustering
df_unsup = X_train_safe.copy()

required_cols = ["LAT", "LON", "occ_hour"]
missing = [c for c in required_cols if c not in df_unsup.columns]
print("Missing required columns:", missing)

# Keep only the needed columns (proposal requirement)
df_unsup = df_unsup[[c for c in required_cols if c in df_unsup.columns]].copy()
print("df_unsup shape (before cleaning):", df_unsup.shape)

df_unsup.head()
```

## Step 2) Clean geo-temporal features

Clustering is very sensitive to noisy/out-of-range values.
We therefore:
- remove rows with missing LAT/LON/HOUR
- keep only plausible ranges:
  - LAT roughly [33, 35], LON roughly [-119, -117] (LA area sanity)
  - occ_hour in [0, 23]

```python
# Drop missing
df_unsup = df_unsup.dropna()

# Clean hour
if "occ_hour" in df_unsup.columns:
    df_unsup = df_unsup[(df_unsup["occ_hour"] >= 0) & (df_unsup["occ_hour"] <= 23)]

# Clean LAT/LON (simple LA-region sanity check)
if "LAT" in df_unsup.columns and "LON" in df_unsup.columns:
    df_unsup = df_unsup[(df_unsup["LAT"] >= 33.0) & (df_unsup["LAT"] <= 35.0)]
    df_unsup = df_unsup[(df_unsup["LON"] >= -119.0) & (df_unsup["LON"] <= -117.0)]

print("df_unsup shape (after cleaning):", df_unsup.shape)
df_unsup.describe()
```

Scale features (required)

K-Means and DBSCAN rely on distances.
LAT/LON and HOUR are in different numeric ranges, so we standardize features.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_unsup_scaled = scaler.fit_transform(df_unsup)

print("X_unsup_scaled shape:", X_unsup_scaled.shape)
```

## K-Means: choosing the number of clusters (K)

K-Means requires selecting K.
We use:
- **Elbow method** (inertia)
- **Silhouette score** (cluster separation quality)

Then we choose a reasonable K that balances interpretability and fit quality.

In the K-Means clustering step, the parameter **K** represents the number of distinct *geospatial–temporal crime profiles* identified in the data.

Each cluster groups crime incidents that are:
- geographically close (similar latitude and longitude),
- and temporally similar (similar hour of occurrence).

Therefore, a cluster does **not** represent a crime type, but rather a recurring **pattern of where and when crimes tend to occur**.  
For example, one cluster may correspond to incidents concentrated in central areas during late-night hours, while another may represent crimes occurring in residential zones during daytime.

Selecting the optimal value of **K** allows us to segment the city into a limited number of interpretable crime profiles, which can be used to identify hot-spots and support resource allocation and crime prevention strategies.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

K_values = list(range(2, 16))
inertias = []
sil_scores = []

# Keep inertia on full sample (ok)
# Compute silhouette on a smaller subset (VERY important for speed)
SIL_SAMPLE_SIZE = 15000  # can be 5000 if still slow

rng = np.random.RandomState(42)
idx_sil = rng.choice(X_sample_scaled.shape[0], size=min(SIL_SAMPLE_SIZE, X_sample_scaled.shape[0]), replace=False)
X_sil = X_sample_scaled[idx_sil]

for k in K_values:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_sample_scaled)
    inertias.append(km.inertia_)

    # silhouette on subset only
    sil_scores.append(silhouette_score(X_sil, km.predict(X_sil)))

# Elbow plot
plt.figure(figsize=(7,4))
plt.plot(K_values, inertias, marker="o")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.title("K-Means Elbow Method")
plt.tight_layout()
plt.show()

# Silhouette plot
plt.figure(figsize=(7,4))
plt.plot(K_values, sil_scores, marker="o")
plt.xlabel("K")
plt.ylabel("Silhouette score (subset)")
plt.title("K-Means Silhouette Scores (computed on subset)")
plt.tight_layout()
plt.show()

# best_k = K_values[int(np.argmax(sil_scores))]
# print("Best K by silhouette (subset suggestion):", best_k)
```

### Choosing the Number of Clusters (K-Means)

We evaluated the number of clusters using both the **Elbow Method** and the
**Silhouette Score**, which provide complementary perspectives.

The Elbow Method analyzes **inertia**, defined as the sum of squared distances
between points and their assigned cluster centroids. Inertia decreases
monotonically as K increases, but the optimal K is identified at the point
where additional clusters yield diminishing returns. In our results, the
inertia curve shows a clear flattening around **K ≈ 6–8**, indicating that
beyond this range, improvements in compactness are marginal.

The Silhouette Score measures how well clusters are separated, with higher
values indicating clearer separation. The silhouette analysis suggests
**K = 2** as the optimal value, corresponding to a coarse partition of the
data into two globally distinct groups.

While K=2 maximizes separation, it is overly simplistic for crime pattern
analysis. Our objective is to identify **distinct spatio-temporal crime
profiles**, not merely to separate high- and low-density regions. Therefore,
we select **K = 8**, which provides a balance between cluster compactness,
interpretability, and actionable insights, in line with the project goals.

## Fit final K-Means and visualize clusters (hotspots/profiles)

We fit K-Means using the selected K and visualize:
- clusters on a LAT/LON scatter plot (geospatial crime profiles)

```python
K_FINAL = 8

kmeans = KMeans(n_clusters=K_FINAL, random_state=42, n_init=10)
k_labels = kmeans.fit_predict(X_sample_scaled)

df_sample["kmeans_cluster"] = k_labels
print(df_sample["kmeans_cluster"].value_counts().sort_index())
```

### K-Means Clustering – Spatial and Temporal Crime Patterns

We applied K-Means clustering on standardized geographic (latitude, longitude) and temporal (hour of occurrence) features to identify crime hotspots and behavioral patterns. The optimal number of clusters was evaluated using both the Elbow method and the Silhouette score on a representative data subset.

While the Silhouette score was maximized at K=2, this solution produced overly coarse clusters that lacked practical interpretability. In contrast, the Elbow method revealed diminishing returns beyond K≈8, suggesting a more meaningful partitioning of the city. Therefore, K=8 was selected as a balanced compromise between cluster compactness and interpretability.

The resulting clusters reveal clear geographic crime regions, each associated with distinct temporal patterns. Some clusters are characterized by early-morning activity, while others peak during evening or nighttime hours. ##  DBSCAN: choosing eps (k-distance plot)

DBSCAN needs:
- eps: neighborhood radius
- min_samples: minimum points to form a dense region

We estimate eps via the **k-distance plot** (k = min_samples).

```python
from sklearn.neighbors import NearestNeighbors

min_samples = 20  # good starting point for large datasets

nbrs = NearestNeighbors(n_neighbors=min_samples).fit(X_sample_scaled)
distances, _ = nbrs.kneighbors(X_sample_scaled)

# k-distance = distance to the kth nearest neighbor
k_dist = np.sort(distances[:, -1])

plt.figure(figsize=(7,4))
plt.plot(k_dist)
plt.xlabel("Points sorted by distance")
plt.ylabel(f"Distance to {min_samples}th nearest neighbor")
plt.title("DBSCAN k-distance plot (choose eps near the elbow)")
plt.tight_layout()
plt.show()

print("Tip: choose eps around the elbow region in the curve.")
```

### DBSCAN Parameter Selection (EPS)

To determine an appropriate value for the neighborhood radius (EPS), we used the k-distance plot based on the distance to the 20th nearest neighbor. The curve remains relatively flat for small distances and exhibits a sharp increase beyond approximately 0.2, indicating a transition from dense regions to sparse outliers.

Based on this elbow point, we selected EPS = 0.2 with min_samples = 20. This choice captures dense crime hotspots while allowing DBSCAN to label isolated incidents as noise, which aligns with the goal of identifying irregularly shaped clusters and anomalies.

## Fit DBSCAN and identify hotspots + anomalies

DBSCAN outputs:
- cluster labels: 0,1,2,...
- noise points labeled as **-1** (these are anomalies/outliers by design)

We visualize clusters and report:
- number of clusters found
- percent of points flagged as noise (anomalies)

```python
from sklearn.cluster import DBSCAN

EPS = 0.2  

db = DBSCAN(eps=EPS, min_samples=min_samples, n_jobs=-1)
db_labels = db.fit_predict(X_sample_scaled)

df_sample["dbscan_cluster"] = db_labels

n_clusters = len(set(db_labels)) - (1 if -1 in db_labels else 0)
noise_pct = (db_labels == -1).mean() * 100

print("DBSCAN clusters found:", n_clusters)
print(f"Noise (anomalies) percentage: {noise_pct:.2f}%")

df_sample["dbscan_cluster"].value_counts().head(10)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df_sample["LON"], df_sample["LAT"], c=df_sample["dbscan_cluster"], s=3)
plt.xlabel("Longitude (LON)")
plt.ylabel("Latitude (LAT)")
plt.title(f"DBSCAN Clusters on Map (eps={EPS}, min_samples={min_samples})")
plt.tight_layout()
plt.show()
```
